```python
from sklearn import datasets
iris = datasets.load_iris()
X = iris.data
```

### 데이터 스케일 맞추기
데이터의 스케일을 맞추면 모형을 최적화하기 쉽다.

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

X_min_max = scaler.fit_transform(X)

from sklearn.cluster import DBSCAN
model_m = DBSCAN(eps=0.3, min_samples=3)
predict_m = model_m.fit_predict(X_min_max)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.scatter(X_min_max[:, 0], X_min_max[:, 1], c=predict_m, s=60, edgecolors='black')
plt.scatter(X_min_max[predict_m==-1,0], X_min_max[predict_m==-1,1], c='red',marker='x')
plt.show()
```
