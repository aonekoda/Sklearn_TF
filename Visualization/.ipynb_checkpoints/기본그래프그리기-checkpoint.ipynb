# 기본 그래프 그리기
___
* 실습에 필요한  라이브러리를 import한다.
* iris dataset을 읽어들인다.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
# Data Loading

iris = pd.read_csv('../data/iris.csv')

print(iris.shape)
iris.head()
```

# 히스토그램
* 도수 분포표를 시각화 한 것.
* 연속형 변수의 구간별 도수를 막대의 길이로 나타내는 그래프
* 데이터 분석 단계에서 변수의 분포, 중심 경향, 퍼짐 정도, 치우침 정도 등을 한눈에 살펴볼 수 있는 시각화 종류로 히스토그램이 많이 사용

### matplotlib에서 히스토그램 그리기
 bins=x 값을 변경해가면서 여러번 시도해 본다.

```python
plt.hist(iris['sepal_width'], bins=10)
plt.show()
```

Y축을 빈도수(frequency)가 아니라 density로 하고 싶을 때는 density=True 를 설정

```python
plt.hist(iris['sepal_width'], bins=10, density=True)
plt.show()
```

### seaborn에서 히스토그램 

```python
sns.distplot(iris['sepal_width'])

plt.show()
```

만약 density 그래프를 표시하고 싶지 않으면 kde=False

```python
sns.distplot(iris['sepal_width'], kde=False)
plt.show()
```

만약 density plot만 그리고 싶다면 hist=False

```python
sns.distplot(iris['sepal_width'], hist=False)
plt.show()
```

histogram에 rug를 같이 표시

```python
sns.distplot(iris['sepal_width'], rug=True)
plt.show()
```

### pandas를 이용한 히스토그램 그리기

```python
iris['sepal_width'].hist(bins=10, grid=False)

plt.show()
```

### 여러개 그룹의 히스토그램, 밀도곡선 그리기
setosa, versicolor, virginica 종의 3개 그룹(groups)의 히스토그램과 밀도곡선을 그룹별로 색깔을 다르게 하여
 표시할 수 있다.

```python
# seaborn 사용하기
sns.distplot(iris[iris.species == "setosa"]["petal_length"], 
             color="blue", label="setosa")

sns.distplot(iris[iris.species == "versicolor"]["petal_length"], 
             color="red", label="versicolor")

sns.distplot(iris[iris.species == "virginica"]["petal_length"], 
             color="green", label="virginica")

plt.legend(title="Species")
plt.show()
```

```python
# pandas 사용하기
iris.groupby('species')['petal_length'].hist(bins=7, grid=False, alpha=0.5)
plt.legend(['setosa','versicolor','virginica'])
```

# 막대 그래프
* tips 데이터셋을 사용하여 막대그래프를 그린다.
* Python의 matplotlib, seaborn, pandas를 사용하여 그리는 방법

```python
tips = pd.read_csv('../data/tips.csv')
print(tips.shape)
tips.head()
```

### matplotlib으로 막대 그리프 그리기
요일(day)별로 tip 의 합계를 구해서 막대 그래프로 비교

```python
tips_sum_by_day = tips.groupby('day').tip.sum()
tips_sum_by_day

label = ['Thur', 'Fri', 'Sat', 'Sun']
index = np.arange(len(label))

plt.bar(index, tips_sum_by_day)
plt.xticks(index, label)
plt.show()
```

**옆으로 누운 막대그래프는 plt.barh() 함수를 사용**

```python
plt.barh(index, tips_sum_by_day)
plt.title('Sum of Tips by Day', fontsize=18)
plt.ylabel('Day', fontsize=15)
plt.xlabel('Sum of Tips', fontsize=15)
plt.yticks(index, label, fontsize=13, rotation=0)
plt.show()
```

### 그룹으로 나누어 비교하는 막대그래프
성별(sex)에 따라 요일별 tip의 평균을 비교한다.

```python
tips_mean_by_day_male = tips[tips['sex'] == 'Male'].groupby('day').tip.mean()
tips_mean_by_day_female = tips[tips['sex'] == 'Female'].groupby('day').tip.mean()

label = ['Thur', 'Fri', 'Sat', 'Sun']
index = np.arange(len(label))

plt.bar(index, tips_mean_by_day_male, width=0.4, label='Male')
plt.bar(index+0.5, tips_mean_by_day_female, width=0.4, label='Female')

plt.xticks(index, label)
plt.legend()
plt.show()
```

### 누적막대그래프로 표시하기

```python
# summary by group

p1 = plt.bar(index, tips_mean_by_day_male, color='b', alpha=0.5, label='Male')
p2 = plt.bar(index, tips_mean_by_day_female, color='r', alpha=0.5,
             bottom=tips_mean_by_day_male, label='Female')

plt.xticks(index, label)
plt.legend()
plt.show()
```

###  seaborn 으로 막대 그래프 그리기

```python
# 요일별 데이터 빈도
sns.countplot('day', data=tips)
plt.title('Seaborn count plot')
plt.show()
```

```python
# 요일별 성별 데이터 빈도
sns.countplot('day', hue='sex', data=tips)
plt.title('Seaborn count plot')
plt.show()
```

```python
# 요일별 성별에 따른 total bill의 평균을 막대 그래프로
sns.barplot(x="day", y="tip", hue="sex", data=tips)
plt.show()
```

# 산점도
iris 데이터를 사용하여 matplotlib, seaborn, pandas에서 산점도를 그려본다.

### Matplotlib를 이용하여 산점도 그리기

```python
plt.plot('petal_length',  # x
         'petal_width',  # y
         data=iris, 
         linestyle='none', 
         marker='o', 
         markersize=10,
         color='blue', 
         alpha=0.5)

plt.title('Scatter Plot of iris by matplotlib', fontsize=20)
plt.xlabel('Petal Length', fontsize=14)
plt.ylabel('Petal Width', fontsize=14)
plt.show()
```

```python
# 그룹별 산점도 그리기 
groups = iris.groupby('species')

fig, ax = plt.subplots()
for name, group in groups:
    ax.plot(group.petal_length, 
            group.petal_width, 
            marker='o', 
            linestyle='',
            label=name)
ax.legend(fontsize=12, loc='upper left') # legend position
```

### seaborn으로 산점도 그리기

```python
# 그룹별 산점도
sns.scatterplot(x='petal_length', 
               y='petal_width', 
                hue='species', # different colors by group
                style='species', # different shapes by group
                s=100, # marker size
                data=iris)
plt.show()
```

### Pandas로 산점도 그리기

```python
ax = groups.get_group('setosa').plot.scatter(x='petal_length', y='petal_width', 
                                                      color='DarkBlue', label='setosa')

groups.get_group('versicolor').plot.scatter(x='petal_length', y='petal_width', 
                                                     color='DarkGreen', label='versicolor', ax=ax)

groups.get_group('virginica').plot.scatter(x='petal_length', y='petal_width', 
                                                     color='Red', label='virginica', ax=ax)
plt.show()
```

# 등고선 그래프 contour

* 등고선 그래프로 3차원 데이터를 시각화 할 수 있다.
* contour는 등고선만 표시하고 contourf는 색깔로 표시한다.

```python
def f(x, y):
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 - y ** 2)


n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
XX, YY = np.meshgrid(x, y)
ZZ = f(XX, YY)

plt.title("Contour plots")
plt.contourf(XX, YY, ZZ, alpha=.75, cmap='jet')
plt.contour(XX, YY, ZZ, colors='black')
plt.show()
```
